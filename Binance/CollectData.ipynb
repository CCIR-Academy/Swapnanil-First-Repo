```python
import time
import logging
from binance.lib.utils import config_logging
from binance.websocket.spot.websocket_client import SpotWebsocketClient as Client
 
config_logging(logging, logging.DEBUG)
 
messages = []

def message_handler(message):
    messages.append(message)
 
 
my_client = Client()
my_client.start()
 
my_client.agg_trade(
    symbol="btcusdt",
    id=1,
    callback=message_handler,
)
 
time.sleep(2)
 
my_client.agg_trade(
    symbol="bnbusdt",
    id=2,
    callback=message_handler,
)
 
time.sleep(60)
 
logging.debug("closing ws connection")
my_client.stop()

# print(messages)
```

```python
for i in range(len(messages)):
    messages[i] = list(messages[i].values())

messages
```

```python
import csv

with open('BinanceDataObtained.csv', 'a') as file:
    writerobj = csv.writer(file)
    writerobj.writerow(['Event type', 'Event time', 'Symbol', 'Aggregate trade ID', 'Price', 'Quantity', 'First trade ID', 'Last trade ID', 'Trade time', 'Is the buyer the market maker?', 'Ignore'])
    writerobj.writerows(messages)
```
